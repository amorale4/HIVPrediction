```python
path_style = 'font-size:12px;fill-rule:nonzero;stroke:#000000;stroke-opacity:1;stroke-width:0.1;stroke-miterlimit:4;stroke-dasharray:none;stroke-linecap:butt;marker-start:none;stroke-linejoin:bevel;fill:'
#colors = ["#F1EEF6", "#D4B9DA", "#C994C7", "#DF65B0", "#DD1C77", "#980043"]
colors = ["#FFF5F0", "#FEE0D2", "#FCBBA1", "#FC9272", "#FB6A4A", "#EF3B2C", "#CB181D", "#99000D"]
```

```python
K = len(colors)
for p in paths:
    #print (p['id'])
    if p['id'] not in ["State_Lines", "separator"]:
        try:
            tweet_count =  numTweets[p['inkscape:label']][0]
        except:
            continue
    
        c = (max_tweet_value - min_tweet_value)/K
        
        if max_tweet_value - c < tweet_count:
            color_class = 7
        elif max_tweet_value - 2*c < tweet_count:
            color_class = 6
        elif max_tweet_value - 3*c < tweet_count:
            color_class = 5
        elif max_tweet_value - 4*c < tweet_count:
            color_class = 4
        elif max_tweet_value - 5*c < tweet_count:
            color_class = 3
        elif max_tweet_value - 6*c < tweet_count:
            color_class = 2
        elif max_tweet_value - 7*c < tweet_count:
            color_class = 1
        else:
            color_class = 0

        color = colors[color_class]
        p['style'] = path_style + color
    
```
